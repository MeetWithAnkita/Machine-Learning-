```python
import numpy as np
from scipy import stats 
```

```python
g1 = [85, 88, 90, 92, 87, 85, 89, 91, 86, 88]
g2 = [82, 84, 80, 83, 81, 79, 78, 85, 84, 83]

g1_std = np.std(g1, ddof = 1)
g2_std = np.std(g2, ddof = 1)

n1 = len(g1)
n2 = len(g2)

mean_g1 = np.mean(g1)
mean_g2 = np.mean(g2)

df = min(n1 - 1, n2 - 1)  # degrees of freedom

```

```python
T_test , p_value = stats.ttest_ind(g1, g2, equal_var = False)
print("T-test value:", T_test)
print("P-Value:", p_value)
```

```python
alpha = 0.05 
if p_value < alpha:
    print("Reject the null hypothesis: The two group means are significantly different.")   
else:
    print("Accept the null hypothesis: The two group means are not significantly different.")
```
