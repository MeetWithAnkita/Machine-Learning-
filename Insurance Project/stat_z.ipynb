```python
import numpy as np
from scipy.stats import norm 
# norm is Normal distribution from scipy.stats library
```

```python
sample_data = [172, 175, 168, 169, 171, 173, 175, 170, 169, 172]
population_mean = 170
population_std = 3
sample_mean = np.mean(sample_data)
n = len(sample_data)
print("Sample Mean:", sample_mean)
print("Number of observations in sample:", n)
```

```python
z_score = (sample_mean - population_mean) / (population_std / np.sqrt(n))
print("Z-Score:", z_score)
```

```python
alpha = 0.05
p_value = 2 * (1 - norm.cdf(abs(z_score)))
# norm.cdf gives cumulative distribution function value for given z_score
print("P-Value:", p_value)
```

```python
# p-value can determine by comparing the significance value (alpha)
if p_value < alpha:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Accept the null hypothesis: The sample mean is not significantly different from the population mean.")
```
